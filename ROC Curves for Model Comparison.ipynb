```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc

# Simulated data (Replace with actual FPR, TPR values from your models)
models = ["Random Forest", "XGBoost", "SVM", "Decision Tree", "Neural Network"]
fpr = {model: np.linspace(0, 1, 100) for model in models}  # False Positive Rate (example)
tpr = {
    "Random Forest": np.sqrt(fpr["Random Forest"]),
    "XGBoost": np.power(fpr["XGBoost"], 0.8),
    "SVM": np.power(fpr["SVM"], 1.1),
    "Decision Tree": np.power(fpr["Decision Tree"], 1.2),
    "Neural Network": np.power(fpr["Neural Network"], 0.9)
}

# Compute AUC values (replace with real data)
auc_values = {model: auc(fpr[model], tpr[model]) for model in models}

# Plot ROC Curves
plt.figure(figsize=(8, 6))
for model in models:
    plt.plot(fpr[model], tpr[model], label=f"{model} (AUC = {auc_values[model]:.2f})")

# Plot random guess line
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random Guess")

# Formatting plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison of Machine Learning Models for NSCLC Prediction")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
